```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mat1 = np.load('data/cleaned_eng.txt.entropy.100000.1000.npy')
X = np.arange(0, 85236506, 100000)
Y_min = mat1[1:,].min(axis=0)
Y_max = mat1[1:,].max(axis=0)
print(mat1[0,-1])
```

```python
plt.figure(figsize=(12, 4))
plt.plot(X, mat1[0,:], label='Original', color='red')
plt.plot(X, mat1.mean(axis=0), label='Mean', color='blue')
plt.fill_between(X, Y_min, Y_max, alpha=0.5)
plt.legend()
plt.ylim(4.4, 4.7)
plt.xlabel('nTokens')
plt.ylabel('Entropy')
plt.title('Entropy of English Tokens')
plt.savefig('entropy_eng.svg')
```
