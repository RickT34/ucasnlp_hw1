```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_entropy_acc(X, mat, ylim, skip_first=False):
    if skip_first:
        Y_min = mat[1:,:].min(axis=0)
        Y_max = mat[1:,:].max(axis=0)
        Y_mean = mat[1:,:].mean(axis=0)
    else:
        Y_min = mat.min(axis=0)
        Y_max = mat.max(axis=0)
        Y_mean = mat.mean(axis=0)
    plt.figure(figsize=(12, 4))
    plt.plot(X, mat[0,:], label='Original', color='red')
    plt.plot(X, Y_mean, label='Mean', color='blue')
    plt.fill_between(X, Y_min, Y_max, alpha=0.5)
    plt.legend()
    plt.ylim(ylim)
    plt.xlabel('nTokens')
    plt.ylabel('Entropy')
```

```python
mat1 = np.load('data/cleaned_eng.txt.entropy_accumulate.100000.1000.npy')
X = np.arange(0, 85236506, 100000)
print(mat1[0,-1])
plot_entropy_acc(X, mat1, (4.4,4.7))
plt.title('Entropy of English Tokens')
plt.savefig('entropy_eng.svg')
```

```python
mat2 = np.load('data/cleaned_chs.txt.entropy_accumulate.100000.1000.npy')
X = np.arange(0, 97395557, 100000)
print(mat2[0,-1])
plot_entropy_acc(X, mat2, (9.5,9.7))
plt.title('Entropy of Chinese Tokens')
plt.savefig('entropy_chs.svg')
```

```python
plot_entropy_acc(X, mat2, (9.5,9.7), skip_first=True)
plt.title('Entropy of Chinese Tokens (without original)')
plt.savefig('entropy_chs2.svg')
```

```python
mat3 = np.load('data/cleaned_eng_lower.txt.entropy_accumulate.100000.1000.npy')
X = np.arange(0, 85236506, 100000)
print(mat3[0,-1])
plot_entropy_acc(X, mat3, (4.16,4.25))
plt.title('Entropy of English (Lowercase) Tokens')
plt.savefig('entropy_englow.svg')
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.set_ylabel('Entropy (Normal)', color='b')
ax1.fill_between(X,mat1.min(axis=0) , mat1.max(axis=0), alpha=0.5, label='Normal', color='b')
ax1.set_ylim(4.4,4.7)
ax2 = ax1.twinx()
ax2.set_ylabel('Entropy (Lowercase)', color='r')
ax2.fill_between(X,mat3.min(axis=0) , mat3.max(axis=0), alpha=0.4, label='Lowercase', color='r')
ax2.set_ylim(4.055,4.355)
ax1.set_xlabel('nTokens')
plt.title('Entropy of Normal and Lowercase Tokens')
plt.savefig('entropy_eng_nvsl.svg')
```
